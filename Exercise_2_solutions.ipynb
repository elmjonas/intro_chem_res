## Exercise 2.2: First fit
Use scipy optimize to fit the function for variables $a$ and $b$ to the data from before.

$$f(x)=\exp\left(x^a+b\right) $$

Start by reusing the code from before, add the fitting function and then plotting it using plt.plot().

Hint: Remember numpy has np.exp() and that more x-points for plotting the fit function can be made with np.arange(1,6,0.01).

```python
import matplotlib.pyplot as plt
import scipy.optimize as sco
import numpy as np
x = [1,2,3,4,5,6]
y = [10,18,21,37,53,99]
###YOUR CODE HERE
def func(x, a, b):
    return np.exp(x**a+b)
(a, b), var = sco.curve_fit(func, x, y)

plt.scatter(x, y, label='Energy', color='red')
fit_x = np.arange(1, 6, 0.01)
plt.plot(fit_x, func(fit_x, a, b), label='Fit', color='blue')

plt.xlabel('Size [Å]')
plt.ylabel('Energy [kcal/mol]')
plt.legend()
plt.show()
###END OF YOUR CODE
```
